```python

print("Insert your card")

Attempts = 0
Max_Attempts = 3
Card_pin = 4325

while Attempts < Max_Attempts:
    Pin = int(input("Enter your four-digit PIN: "))

    if Card_pin == Pin:
        print("PIN correct")
        print("You are Welcome to our bank")

        transaction_choice = input("Enter your transaction choice (B, D, W, C): ").upper()

        if transaction_choice == "B":
            print("Your Balance is ₦10.00")

        elif transaction_choice == "D":
            print("You want to Deposit")

        elif transaction_choice == "W":
            print("You are making a Withdrawal")

        elif transaction_choice == "C":
            print("Transaction Cancelled")

        else:
            print("Invalid Transaction Choice")

        break

    else:
        Attempts += 1
        print("Incorrect PIN, You are Tinubu")

        if Attempts < Max_Attempts:
            print(f"You have {Max_Attempts - Attempts} attempt(s) remaining.")
        else:
            print("You have exceeded the maximum number of attempts.")
            print("Access denied.")
```
